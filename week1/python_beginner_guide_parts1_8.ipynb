## Part 2: Variables and Data Types

```
x = 10
name = 'Karan'
pi = 3.14
print(x, name, pi)
```

## Part 3: Control Flow

```
age = 18
if age >= 18:
    print('Adult')
else:
    print('Minor')
```

## Part 4: Loops

```
for i in range(5):
    print(i)
```

## Part 5: Functions

```
def greet(name):
    return f'Hello, {name}'
print(greet('Karan'))
```

## Part 6: Lists and Dictionaries

```
nums = [1,2,3]
info = {'name':'Karan', 'age':25}
print(nums, info)
```

## Part 7: File Handling

```
with open('sample.txt','w') as f:
    f.write('Hello World')
```
