# Part 8: Understanding Python Import Structure

This section explains how imports work in Python, especially in modules and notebooks.

## 1. What is an Import?

An import allows you to use code from another file (module) or library.

Example:

````
import math
print(math.sqrt(16))
````

## 2. Importing Specific Items

Instead of importing everything, you can import specific functions:

````
from math import sqrt
print(sqrt(25))
````

## 3. Aliases

You can rename modules using `as`:

````
import numpy as np
arr = np.array([1,2,3])
print(arr)
````

## 4. Importing Your Own Modules

If you have a file `my_module.py`:

````
# my_module.py

def greet(name):
    return f"Hello, {name}!"
````

````
import my_module
print(my_module.greet("Karan"))
````

## 5. Folder Structure (Packages)

Example project structure:

```
project/
│── main.py
│── utils/
│   ├── __init__.py
│   └── helper.py
```

Import like:

````
from utils.helper import some_function
````

## 6. Relative vs Absolute Imports

Absolute import:

````
from utils.helper import some_function
````

Relative import (inside packages):

````
from .helper import some_function
````

## 7. sys.path and Import Resolution

Python searches for modules in paths listed in `sys.path`:

````
import sys
print(sys.path)
````
